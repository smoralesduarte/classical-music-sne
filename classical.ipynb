```python
import pandas as pd
import numpy as np
# import requests
# import zipfile
# import io
from openTSNE import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt # Still used for k-distance plot
import plotly.graph_objects as go # Import Plotly
import os
import warnings
import re

# Suppress specific pandas warnings if needed (optional)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)


# --- 1. Data Retrieval (Load Local Spotify Tracks CSV) ---
local_csv_file_path = 'dataset.csv' # Make sure this file is downloaded from Kaggle
track_df = None # Initialize
print(f"Attempting to load data from local file: {local_csv_file_path}")

if os.path.exists(local_csv_file_path):
    try:
        track_df = pd.read_csv(local_csv_file_path)
        if 'Unnamed: 0' in track_df.columns:
            track_df = track_df.drop(columns=['Unnamed: 0'])
        print(f"Loaded {len(track_df)} tracks from local CSV.")

    except Exception as e:
        print(f"An unexpected error occurred during local file loading: {e}")
        track_df = None # Ensure df is None on error
else:
    print(f"Error: Local file not found at '{local_csv_file_path}'. Please download it first from Kaggle.")


if track_df is not None:
    # --- 2. Filter for Classical Music ---
    print("Filtering for classical music tracks...")
    classical_genres_pattern = r'classical|baroque|romantic|impressionist|opera|choral|chamber|symphony|orchestra|sonata|concerto|cantata|piano|violin|cello|harpsichord|string quartet'
    if 'track_genre' in track_df.columns:
        track_df['track_genre'] = track_df['track_genre'].astype(str)
        classical_mask = track_df['track_genre'].str.contains(classical_genres_pattern, case=False, regex=True, na=False)
        classical_df = track_df[classical_mask].copy()
        print(f"Found {len(classical_df)} tracks matching classical genres.")
    else:
        print("Error: 'track_genre' column not found in the dataset.")
        classical_df = pd.DataFrame()

    if not classical_df.empty:
        # --- 3. Select and Prepare Features ---
        feature_cols = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        missing_cols = [col for col in feature_cols if col not in classical_df.columns]
        if missing_cols:
            print(f"Error: Missing required feature columns: {missing_cols}")
            classical_df = pd.DataFrame()
        else:
            print(f"Selected features: {feature_cols}")
            initial_count = len(classical_df)
            classical_df.dropna(subset=feature_cols + ['artists'], inplace=True)
            if len(classical_df) < initial_count:
                 print(f"Dropped {initial_count - len(classical_df)} rows with NaNs in features or artists.")
            print(f"Number of classical tracks after dropping NaNs: {len(classical_df)}")
            if classical_df.empty:
                 print("Error: No valid classical tracks remaining after cleaning.")

    if not classical_df.empty:
        # --- 4. Normalize Features ---
        print("Normalizing audio features...")
        scaler = StandardScaler()
        classical_df.loc[:, feature_cols] = scaler.fit_transform(classical_df[feature_cols])

        # --- 5. Group by Artist and Aggregate Features ---
        print("Grouping by artist and aggregating features...")
        classical_df['main_artist'] = classical_df['artists'].astype(str).apply(lambda x: x.split(';')[0].strip())
        artist_song_counts = classical_df['main_artist'].value_counts()
        print(f"Calculated song counts for {len(artist_song_counts)} unique artists.")
        artist_aggregated_features = classical_df.groupby('main_artist')[feature_cols].mean()

        if len(artist_aggregated_features) < 10:
             print(f"Warning: Only {len(artist_aggregated_features)} unique artists found after filtering. Results may be sparse.")
        print(f"Aggregated features for {len(artist_aggregated_features)} unique classical artists/composers.")

        artist_names = artist_aggregated_features.index.tolist()
        artist_features_np = artist_aggregated_features.values.astype(np.float32)

        # --- 6. Experiment Loop: t-SNE + DBSCAN on Aggregated Features ---
        N_COMPONENTS = 3
        alpha_values = [0.6, 0.7, 0.8, 0.9, 1.0]
        eps_map = {0.6: 1.5, 0.7: 4.0, 0.8: 4.5, 0.9: 5.0, 1.0: 5.5}
        dbscan_min_samples = 2 * N_COMPONENTS

        print(f"\n--- Starting Experiment Loop with alpha in (0.5, 1.0] ---")
        print(f"Alpha values to test: {alpha_values}")
        print(f"Corresponding eps values to use: {eps_map}")
        print(f"Using DBSCAN min_samples = {dbscan_min_samples}")
        print("Note: k-distance plots still generated for validation.")

        results = {}

        for alpha in alpha_values:
            current_eps = eps_map[alpha]
            print(f"\n--- Processing alpha = {alpha}, using eps = {current_eps} ---")
            dof = max(0.01, 2 * alpha - 1)
            print(f"Using alpha = {alpha}, corresponding to approx. dof = {dof:.2f}")

            print(f"Running OpenTSNE on {len(artist_aggregated_features)} artists...")
            perplexity_val = min(30, len(artist_aggregated_features) - 2)
            if perplexity_val < 5: perplexity_val = 5

            tsne_3d = TSNE(
                n_components=N_COMPONENTS, perplexity=perplexity_val, metric="euclidean",
                dof=dof, n_jobs=-1, random_state=42, verbose=False, n_iter=750
            )
            artist_embedding_3d = tsne_3d.fit(artist_features_np)
            print("OpenTSNE complete.")

            # --- Calculate and Plot k-distance graph (still useful for validation) ---
            print(f"Calculating k-distance graph for validation (k={dbscan_min_samples})...")
            neighbors = NearestNeighbors(n_neighbors=dbscan_min_samples)
            neighbors_fit = neighbors.fit(artist_embedding_3d)
            distances, indices = neighbors_fit.kneighbors(artist_embedding_3d)
            k_distances = np.sort(distances[:, dbscan_min_samples-1], axis=0)

            plt.figure(figsize=(8, 4))
            plt.plot(k_distances)
            plt.title(f'k-Distance Graph (k={dbscan_min_samples}) for alpha={alpha}')
            plt.xlabel("Points sorted by distance")
            plt.ylabel(f"{dbscan_min_samples}-th Nearest Neighbor Distance")
            plt.axhline(y=current_eps, color='r', linestyle='--', label=f'Used eps = {current_eps}')
            plt.legend()
            plt.grid(True)
            plt.show() # Show matplotlib k-distance plot
            print(f"--> K-distance plot generated. Check if the chosen eps={current_eps} aligns with an 'elbow'.")

            # --- Run DBSCAN Clustering ---
            print(f"Running DBSCAN with eps={current_eps} and min_samples={dbscan_min_samples}...")
            dbscan = DBSCAN(eps=current_eps, min_samples=dbscan_min_samples, metric='euclidean')
            clusters = dbscan.fit_predict(artist_embedding_3d)
            n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
            n_noise = np.sum(clusters == -1)
            print(f"DBSCAN found {n_clusters} clusters and {n_noise} noise points.")

            results[alpha] = {
                'embedding': artist_embedding_3d, 'clusters': clusters,
                'n_clusters': n_clusters, 'n_noise': n_noise,
                'dof': dof, 'eps_used': current_eps
            }

            # --- Output Artists per Cluster (Sorted by Song Count) ---
            print("\n--- Sample Artists per Cluster (Sorted by Song Count Desc) ---")
            unique_clusters_sorted = sorted([c for c in np.unique(clusters) if c != -1])
            # Prepare data for Plotly hover text
            plot_labels = []
            for i in range(len(artist_names)):
                artist = artist_names[i]
                cluster = clusters[i]
                count = artist_song_counts.get(artist, 0)
                label_text = f"Artist: {artist}<br>Cluster: {'Noise' if cluster == -1 else cluster}<br>Songs: {count}"
                plot_labels.append(label_text)

            if not unique_clusters_sorted:
                print("No clusters found (only noise).")
            else:
                for cluster_label in unique_clusters_sorted:
                    cluster_indices = np.where(clusters == cluster_label)[0]
                    cluster_artists = [artist_names[i] for i in cluster_indices]
                    sorted_cluster_artists = sorted(
                        cluster_artists, key=lambda artist: artist_song_counts.get(artist, 0), reverse=True
                    )
                    print(f"\nCluster {cluster_label} ({len(cluster_artists)} artists):")
                    print(f"  Sample (Top Artists by Song Count): {', '.join(sorted_cluster_artists[:10])}")
                    if len(sorted_cluster_artists) > 10: print("  ...")

            print(f"\nNoise points (-1): {n_noise} artists")
            noise_indices = np.where(clusters == -1)[0]
            if n_noise > 0:
                 noise_artists = [artist_names[i] for i in noise_indices]
                 sorted_noise_artists = sorted(
                     noise_artists, key=lambda artist: artist_song_counts.get(artist, 0), reverse=True
                 )
                 if n_noise <= 20:
                     print(f"  Sample Noise (Top Artists by Song Count): {', '.join(sorted_noise_artists[:10])}{'...' if len(sorted_noise_artists) > 10 else ''}")
                 else:
                     print(f"  Sample Noise (Top Artists by Song Count): {', '.join(sorted_noise_artists[:10])}...")
                     print("  (Too many noise artists to list all)")
            print("-" * 30)


            # --- Visualization with Plotly (Interactive 3D) ---
            print("Generating interactive 3D plot with Plotly...")

            fig = go.Figure()

            # Add trace for noise points first (if any)
            noise_mask = clusters == -1
            if np.sum(noise_mask) > 0:
                fig.add_trace(go.Scatter3d(
                    x=artist_embedding_3d[noise_mask, 0],
                    y=artist_embedding_3d[noise_mask, 1],
                    z=artist_embedding_3d[noise_mask, 2],
                    mode='markers',
                    marker=dict(
                        size=3,
                        color='grey', # Assign grey color to noise
                        opacity=0.5,
                        symbol='x'    # Use 'x' marker for noise
                    ),
                    text=[plot_labels[i] for i in noise_indices], # Hover text for noise
                    hoverinfo='text',
                    name='Noise'      # Legend entry for noise
                ))

            # Add trace for each cluster
            # Use Plotly's default color cycle by adding traces per cluster
            for cluster_label in unique_clusters_sorted:
                cluster_mask = clusters == cluster_label
                cluster_indices_for_text = np.where(cluster_mask)[0]
                fig.add_trace(go.Scatter3d(
                    x=artist_embedding_3d[cluster_mask, 0],
                    y=artist_embedding_3d[cluster_mask, 1],
                    z=artist_embedding_3d[cluster_mask, 2],
                    mode='markers',
                    marker=dict(
                        size=5,  # Slightly larger markers for clusters
                        # color=cluster_label, # Let Plotly assign color based on trace
                        # colorscale='Viridis', # Can specify if needed
                        opacity=0.8
                    ),
                    text=[plot_labels[i] for i in cluster_indices_for_text], # Hover text for clusters
                    hoverinfo='text',
                    name=f'Cluster {cluster_label}' # Legend entry for cluster
                ))

            # Update layout
            fig.update_layout(
                title=f'Interactive 3D Plot: Classical Artists (alpha={alpha}, eps={current_eps})<br>{n_clusters} Clusters, {n_noise} Noise Points',
                scene=dict(
                    xaxis_title='Dimension 1',
                    yaxis_title='Dimension 2',
                    zaxis_title='Dimension 3'
                ),
                margin=dict(l=0, r=0, b=0, t=40), # Adjust margins
                legend_title_text='Clusters'
            )

            fig.show() # Display the interactive plot

        print("\n--- Experiment Loop Finished ---")

    else:
       print("\nNo classical tracks found or remaining after filtering/cleaning. Cannot proceed.")

else:
    print("\nCould not proceed with analysis due to data loading errors.")
```

```python
import pandas as pd
import numpy as np
# import requests # Not needed - loading local file
# import zipfile # Not needed - loading local csv
# import io # Not needed
from openTSNE import TSNE # Use OpenTSNE
# from sklearn.cluster import DBSCAN # REMOVED
from sklearn.preprocessing import StandardScaler # To normalize features
# from sklearn.neighbors import NearestNeighbors # REMOVED
import matplotlib.pyplot as plt # No longer needed for main plots
import plotly.graph_objects as go # Used indirectly via px or for fine control
import plotly.express as px # For easier color mapping
import os # To check if file exists
import warnings
import re # For genre filtering

# Suppress specific pandas warnings if needed (optional)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)


# --- 1. Data Retrieval (Load Local Spotify Tracks CSV) ---
local_csv_file_path = 'dataset.csv' # Make sure this file is downloaded from Kaggle
track_df = None # Initialize
print(f"Attempting to load data from local file: {local_csv_file_path}")

if os.path.exists(local_csv_file_path):
    try:
        track_df = pd.read_csv(local_csv_file_path)
        if 'Unnamed: 0' in track_df.columns:
            track_df = track_df.drop(columns=['Unnamed: 0'])
        print(f"Loaded {len(track_df)} tracks from local CSV.")
        # print("Available columns:", track_df.columns.tolist()) # Optional: Show columns

    except Exception as e:
        print(f"An unexpected error occurred during local file loading: {e}")
        track_df = None # Ensure df is None on error
else:
    print(f"Error: Local file not found at '{local_csv_file_path}'. Please download it first from Kaggle.")


if track_df is not None:
    # --- 2. Genre Selection and Sampling ---
    print("\nInspecting available genres...")
    if 'track_genre' not in track_df.columns:
        print("Error: 'track_genre' column not found. Cannot proceed with genre experiment.")
        track_df = None # Stop processing
    else:
        track_df['track_genre'] = track_df['track_genre'].astype(str).str.lower() # Standardize genre names
        print("Top 50 Genres Found:")
        # Uncomment the next line temporarily to see available genres if needed
        # print(track_df['track_genre'].value_counts().head(50))

        # !!! CHOOSE 5 DISTINCT GENRES HERE based on the actual genres in your CSV !!!
        target_genres = ['classical', 'pop', 'metal', 'jazz', 'techno'] # Example selection - CHANGE IF NEEDED
        print(f"\nSelected target genres: {target_genres}")

        # Filter for tracks belonging *only* to one of the target genres
        genre_filtered_df = track_df[track_df['track_genre'].isin(target_genres)].copy()

        # Sample N tracks per genre
        n_samples_per_genre = 200
        print(f"Attempting to sample {n_samples_per_genre} tracks per selected genre...")
        sampled_dfs = []
        available_genres_in_sample = []
        for genre in target_genres:
            genre_tracks = genre_filtered_df[genre_filtered_df['track_genre'] == genre]
            if len(genre_tracks) == 0:
                 print(f"Warning: No tracks found for genre '{genre}'. Skipping.")
                 continue
            elif len(genre_tracks) < n_samples_per_genre:
                 print(f"Warning: Genre '{genre}' has only {len(genre_tracks)} tracks. Using all available.")
                 sampled_dfs.append(genre_tracks)
                 available_genres_in_sample.append(genre)
            else:
                 sampled_dfs.append(genre_tracks.sample(n=n_samples_per_genre, random_state=42))
                 available_genres_in_sample.append(genre)

        if not sampled_dfs:
             print("Error: No tracks found for any of the selected genres. Cannot proceed.")
             sampled_df = pd.DataFrame() # Empty df
        else:
             sampled_df = pd.concat(sampled_dfs).reset_index(drop=True)
             print(f"\nCreated sampled dataset with {len(sampled_df)} tracks from {len(available_genres_in_sample)} genres.")
             print("Genre distribution in sample:")
             print(sampled_df['track_genre'].value_counts())

    # --- 3. Select and Prepare Features (for sampled tracks) ---
    if not sampled_df.empty:
        feature_cols = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        print(f"\nSelected features: {feature_cols}")
        missing_cols = [col for col in feature_cols if col not in sampled_df.columns]
        if missing_cols:
            print(f"Error: Missing required feature columns: {missing_cols}")
            sampled_df = pd.DataFrame()
        else:
            initial_count = len(sampled_df)
            sampled_df.dropna(subset=feature_cols, inplace=True)
            if len(sampled_df) < initial_count:
                print(f"Dropped {initial_count - len(sampled_df)} rows with NaNs in features.")
            if sampled_df.empty:
                print("Error: No valid tracks remaining after cleaning NaNs.")

    if not sampled_df.empty:
        # --- 4. Normalize Features ---
        print("Normalizing audio features for sampled tracks...")
        scaler = StandardScaler()
        normalized_features = scaler.fit_transform(sampled_df[feature_cols])
        track_features_np = normalized_features.astype(np.float32)

        # Store info needed for plotting/interpretation
        # Make sure index is reset if previous operations messed it up
        sampled_track_info = sampled_df[['track_id', 'track_name', 'artists', 'track_genre']].reset_index(drop=True)
        print(f"Prepared normalized features for {len(track_features_np)} tracks.")


        # --- 5. Experiment Loop: t-SNE (3D & 2D) on Sampled Track Features ---
        # !!! Using extended alpha values as requested !!!
        alpha_values = [0.6, 0.7, 0.8, 0.9, 1.0, 1.5, 2.0, 4.0]

        print(f"\n--- Starting Experiment Loop: Genre Separation (3D vs 2D) ---")
        print(f"Alpha values to test: {alpha_values}")

        results = {} # Still store results if needed later

        for alpha in alpha_values:
            print(f"\n--- Processing alpha = {alpha} ---")
            dof = max(0.01, 2 * alpha - 1)
            print(f"Using alpha = {alpha}, corresponding to approx. dof = {dof:.2f}")

            # --- Run t-SNE (3D) ---
            print(f"Running OpenTSNE (3D) on {len(track_features_np)} tracks...")
            perplexity_val = min(30, len(track_features_np) - 2)
            if perplexity_val < 5: perplexity_val = 5

            tsne_3d_model = TSNE(
                n_components=3, perplexity=perplexity_val, metric="euclidean",
                dof=dof, n_jobs=-1, random_state=42, verbose=False, n_iter=750
            )
            track_embedding_3d = tsne_3d_model.fit(track_features_np)
            print("OpenTSNE (3D) complete.")

            # --- Run t-SNE (2D) ---
            print(f"Running OpenTSNE (2D) on {len(track_features_np)} tracks...")
            tsne_2d_model = TSNE(
                n_components=2, perplexity=perplexity_val, metric="euclidean",
                dof=dof, n_jobs=-1, random_state=42, verbose=False, n_iter=750
            )
            track_embedding_2d = tsne_2d_model.fit(track_features_np)
            print("OpenTSNE (2D) complete.")

            # --- Create DataFrame for Plotting ---
            plot_df = pd.DataFrame({
                'x_3d': track_embedding_3d[:, 0],
                'y_3d': track_embedding_3d[:, 1],
                'z_3d': track_embedding_3d[:, 2],
                'x_2d': track_embedding_2d[:, 0],
                'y_2d': track_embedding_2d[:, 1],
            })
            # Add necessary info - ensure alignment using index if needed, but should be ok here
            plot_df['true_genre'] = sampled_track_info['track_genre']
            plot_df['track_name'] = sampled_track_info['track_name']
            plot_df['artists'] = sampled_track_info['artists']

            # Create simplified hover text (no cluster info)
            hover_text = plot_df.apply(
                lambda row: f"Track: {row['track_name']}<br>Artist: {row['artists']}<br>Genre: {row['true_genre']}",
                axis=1
            )
            plot_df['hover_text'] = hover_text

            # Store results (optional, without DBSCAN info now)
            results[alpha] = {
                'embedding_3d': track_embedding_3d,
                'embedding_2d': track_embedding_2d,
                'dof': dof
            }


            # --- Visualization 1: 3D Plot Colored by TRUE GENRE ---
            print("Generating interactive 3D plot (Colored by True Genre)...")
            fig_genre_3d = px.scatter_3d(
                plot_df,
                x='x_3d', y='y_3d', z='z_3d',
                color='true_genre', # Color points by the original genre
                title=f'3D Track Embedding Colored by TRUE GENRE (alpha={alpha})',
                labels={'color': 'Genre', 'x_3d':'Dim 1', 'y_3d':'Dim 2', 'z_3d':'Dim 3'},
                hover_data={'x_3d':False, 'y_3d':False, 'z_3d':False, 'true_genre':True}, # Customize hover
                custom_data=['hover_text'] # Use this for full hover text
            )
            fig_genre_3d.update_traces(marker=dict(size=3, opacity=0.8), hovertemplate='%{customdata[0]}<extra></extra>')
            fig_genre_3d.update_layout(
                 scene=dict(xaxis_title='Dim 1', yaxis_title='Dim 2', zaxis_title='Dim 3'),
                 margin=dict(l=0, r=0, b=0, t=40),
                 legend_title_text='True Genre'
            )
            fig_genre_3d.show()

            # --- Visualization 2: 2D Plot Colored by TRUE GENRE ---
            print("Generating interactive 2D plot (Colored by True Genre)...")
            fig_genre_2d = px.scatter(
                plot_df,
                x='x_2d', y='y_2d',
                color='true_genre', # Color points by the original genre
                title=f'2D Track Embedding Colored by TRUE GENRE (alpha={alpha})',
                labels={'color': 'Genre', 'x_2d':'Dim 1', 'y_2d':'Dim 2'},
                hover_data={'x_2d':False, 'y_2d':False, 'true_genre':True}, # Customize hover
                custom_data=['hover_text'] # Use this for full hover text
            )
            fig_genre_2d.update_traces(marker=dict(size=5, opacity=0.8), hovertemplate='%{customdata[0]}<extra></extra>') # Slightly larger markers for 2D
            fig_genre_2d.update_layout(
                 margin=dict(l=0, r=0, b=0, t=40),
                 legend_title_text='True Genre'
            )
            fig_genre_2d.show()

            print(f"\n--> Compare the 3D and 2D plots above for alpha={alpha}. How well are genres separated visually?")
            print("-" * 30)


        print("\n--- Experiment Loop Finished ---")

    else:
       print("\nNo valid tracks remaining after filtering/cleaning. Cannot proceed.")

else:
    print("\nCould not proceed with analysis due to data loading errors.")
```
